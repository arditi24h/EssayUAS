```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Membaca file CSV
df = pd.read_csv('Hospitals.csv')
# Menampilkan 5 baris pertama
print(df.head())
# Menampilkan info kolom dan tipe data
print(df.info())
# Menampilkan ringkasan statistik (untuk data numerik)
print(df.describe())
```

```python
sns.histplot(df['Number of Beds'], kde=True)
plt.title('Distribusi Jumlah Tempat Tidur')
plt.xlabel('Jumlah Tempat Tidur')
plt.ylabel('Frekuensi')
plt.show()
```
